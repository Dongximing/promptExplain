```python
import pandas as pd

def classify_instruction_query(df, col_name):
    # This function assumes the input DataFrame and column name are valid.
    def classify(row):
        tokens = row[col_name]['tokens']
        print("ins:", tokens.get('instruction'), "que=", tokens.get('query'))
        return 'instruction' if tokens.get('instruction', 0) > tokens.get('query', 0) else 'query'

    # Apply the function to each row and return the result as a new pandas Series.
    return df.apply(classify, axis=1)


df['inst_or_query_win'] = classify_instruction_query(df, 'component_level')
df['inst_or_query_win']
```

```python
def clean_label(text_list):
    text = ''.join(text_list).upper()  # Convert list to string and uppercase
    if 'POSITIVE' in text:
        return 1  # Positive label
    elif 'NEGATIVE' in text:
        return 0  # Negative label
    else:
        return -1  # Other

for col in cols:
    df[f'label_{col}'] = df[col].apply(clean_label)

```

```python
lebel_cols = []
for col in df.columns:
    if 'label' in col:
        lebel_cols.append(col)

df[lebel_cols].head(10)
```

```python
data = df[lebel_cols]
```

```python
from sklearn.metrics import accuracy_score

# Compute the accuracy between 'label' and 'label_real_output'
accuracy = accuracy_score(data['label'], data['label_real_output'])

# Filter the dataframe based on the 'label_real_output' column
filtered_data = data[data['label_real_output'].isin([0, 1])]

# Initialize an empty dictionary to store the flip rate for each column
flip_rates = {}

# Iterate through each column (excluding 'label' and 'label_real_output') to calculate the flip rate
for column in filtered_data.columns:
    if column not in ['label', 'label_real_output']:
        # Calculate the flip rate as the fraction of values that are different from 'label'
        flip_rate = (filtered_data['label'] != filtered_data[column]).mean()
        flip_rates[column] = flip_rate

# Convert the flip rates dictionary to a dataframe for easier viewing and saving
flip_rate_df = pd.DataFrame(list(flip_rates.items()), columns=['Column', 'Flip Rate'])


print(accuracy)
flip_rate_df
```
